```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.autograd
import torchvision
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
```

```python
def extract_numbers(s):
    if isinstance(s, str):
        return float(re.search(r'\d+\.\d+', s).group())
    else:
        return s
```

```python
csv_control = "/home/obasit/50024/checkpoint3/reweighting-DL/accuracy_tsting_no_weights.csv"
csv_control_df = pd.read_csv(csv_control, index_col=0)
csv_control_df = csv_control_df.map(extract_numbers)
csv_control_means = csv_control_df.mean()
csv_control_std_devs = csv_control_df.std()
```

```python
# Plot
plt.figure(figsize=(8, 6))
plt.errorbar(csv_control_means.index, csv_control_means.values, yerr=csv_control_std_devs.values, fmt='o', linestyle='-', capsize=5)
plt.xlabel('% of minority class')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
